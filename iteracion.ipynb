Ejercicio 1: Suma acumulada de una lista. Dada 
una lista de números, recorre la lista con un bucle y guarda en memoria 
la suma acumulada de los elementos. Al final, muestra la suma total.


```python
def cumulative_sum(numbers):
    sum = 0
    for num in numbers:
        sum = sum + num        
    
    return sum
```

```python
cumulative_sum([2,3,4,5])
```

Ejercicio 2: Encuentra el valor máximo
Recorre una lista de números y almacena en memoria el mayor valor encontrado hasta el momento. Al terminar, imprime el valor máximo.

```python
def max_value(numbers):
    max = 0
    for num in numbers:
        if num > max:
            max = num
    return max
```

```python
max_value([1,4,5,3,12,6])
```

Ejercicio 3: Contar elementos repetidos
Dada una lista de palabras, recórrela e identifica cuántas veces aparece cada palabra usando un diccionario para “memorizar” las apariciones.

```python
def repeat_words(words):
    word_list = {}
    for item in words:
        if item in word_list:
            word_list[item] = word_list[item]+1
        else:
            word_list[item] = 1

    return word_list
```

```python
repeat_words(['hola', 'keep', 'keep', 'hola', 'hola', 'Herminio'])
```

Ejercicio 4: Detectar si hay elementos duplicados
Recorre una lista y utiliza una estructura de memoria (como un set) para guardar los elementos ya vistos. 
Si encuentras uno repetido, muestra un mensaje indicando que hay duplicados.

```python
def duplicate_element(numbers):
    num_list = []
    result = 'Ningún numero duplicado'
    for num in numbers:
        if num in num_list:
            result =  num , 'número duplicado'
            break
        else:
            num_list.append(num)
    return result
    
```

```python
duplicate_element([2,3,3,4,4])
```
